Question 2 : Explain the difference between List and Tuple in Python.

Ans:the difference between list and tuple are

| Feature    | List                          | Tuple                            |
| ---------- | ----------------------------- | -------------------------------- |
| Definition | Ordered collection of items   | Ordered collection of items      |
| Mutability | Mutable (change ho sakta hai) | Immutable (change nahi ho sakta) |
| Syntax     | Square brackets `[ ]`         | Parentheses `( )`                |
| Example    | `[1,2,3]`                     | `(1,2,3)`                        |

EXAMPLE:

# List
my_list = [1, 2, 3]
my_list[0] = 10   # Change possible

# Tuple
my_tuple = (1, 2, 3)
# my_tuple[0] = 10   # Error because tuple immutable hai
